```python
from ghmm import *
import pickle
```

```python
with open("train_skill_5.pkl", "rb") as pkl_file:
    seq_train = pickle.load(pkl_file)
    
with open("test_skill_5.pkl", "rb") as pkl_file:
    seq_test = pickle.load(pkl_file)
```

```python
pi = [0.26, 0.74]
A = [[1, 0], [0.17, 0.83]]
B = [[0.7, 0.3], [0.13, 0.87]]
```

```python
obs = Alphabet([1, 2])
seq_set = SequenceSet(obs, seq_train)
m = HMMFromMatrices(obs, DiscreteDistribution(obs), A, B, pi)
m.baumWelch(seq_set)
```

```python
seq_set = SequenceSet(obs, seq_test)
v = m.viterbi(seq_set)
print v
```
