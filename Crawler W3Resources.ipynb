```python
from bs4 import BeautifulSoup
import requests
from db import PythonProblems
from datetime import datetime
```

```python
url = "https://www.w3resource.com/python-exercises/python-basic-exercises.php"
response = requests.get(url)
data = response.text
soup = BeautifulSoup(data, 'html5lib')
```

```python
exercises = soup.find_all("strong")
problem_list = []

for e in exercises:
    problem_text = ""
    problem_text = e.parent.text
    if e.parent.next_sibling and e.parent.next_sibling.name == "pre":
        problem_text += e.parent.findNext().text
    problem_list.append(problem_text)
```

```python
for item in problem_list:
    print(item)
```
