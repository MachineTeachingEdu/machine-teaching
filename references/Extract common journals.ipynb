```python
import pandas as pd
import qgrid
```

```python
df = pd.read_csv('classificações_publicadas_todas_as_areas_avaliacao1522078273541.xls', encoding='latin1',
                sep='\t')
```

```python
# Filter journals of interest by area
education = df[df['Área de Avaliação'].str.strip() == 'EDUCAÇÃO']
cs = df[(df['Área de Avaliação'].str.strip() == 'CIÊNCIA DA COMPUTAÇÃO') | (df['Área de Avaliação'].str.strip() == 'ENGENHARIAS IV')]
```

```python
# Filter journals of interest by importance
education_plus = education[education['Estrato'].isin(['A1', 'A2', 'B1'])]
cs_plus = cs[cs['Estrato'].isin(['A1', 'A2', 'B1'])]
```

```python
# Get journals that are in both DF
journals = pd.merge(education_plus, cs_plus, on=['Título'], how='inner')
```

```python
qgrid.show_grid(journals)
```

```python
journals.to_excel('common_journals.xlsx')
education = education.applymap(lambda x: x.encode('unicode_escape').
                 decode('utf-8') if isinstance(x, str) else x)
education.to_excel('education.xlsx', encoding='utf8')
cs.to_excel('cs.xlsx')
education_plus.to_excel('education_plus.xlsx')
cs_plus.to_excel('cs_plus.xlsx')
```
